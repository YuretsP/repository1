```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
```

```python
#-----A-----
#-----A-----
#-----A-----
#-----A-----
#-----A-----
x = np.loadtxt('C:\data\A.txt', usecols=0)
y = np.loadtxt('C:\data\A.txt', usecols=1)
```

```python
plt.figure(figsize=(15, 8))
plt.plot(x, y, color='red', label = 'давление 2.7E-1 mbar, ток 0.03 А, напряжение 964.4 V, щель 100 mkm')
plt.ylabel('E, см^(-1)')
plt.xlabel('$\lambda$, нм')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(15, 8))
    #plt.plot(x,y, color='red', label = 'давление 2.7E-1 mbar, ток 0.03 А, напряжение 964.4 V, щель 100 mkm')
plt.ylabel('log(E)')
plt.xlabel('$\lambda$, нм')
plt.xlim(333, 337)
    #change limits
y1 = np.log(y)
plt.plot(x, y1, color='red', label = 'давление 2.7E-1 mbar, ток 0.03 А, напряжение 964.4 V, щель 100 mkm')
plt.legend()
plt.show()
```

```python
x = np.loadtxt('C:\data\A1.txt', usecols=0)
y = np.loadtxt('C:\data\A1.txt', usecols=1)
popt, pcov = curve_fit(lambda t, a, b, c: a*np.exp(b*t)+c,  x-335, y)

a = popt[0]
b = popt[1]
c = popt[2]

plt.figure(figsize=(15, 8))
plt.plot(x,y, color='red', label = 'давление 2.7E-1 mbar, ток 0.03 А, напряжение 964.4 V, щель 100 mkm')
plt.xlabel('$\lambda$, нм')
plt.ylabel('log(E)')
plt.plot(x, a*np.exp(b*(x-335))+c, color='navy', label='fitting curve')
plt.legend()
plt.show()
a, b, c
```
